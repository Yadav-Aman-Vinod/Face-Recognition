```python
import threading
import cv2
from deepface import DeepFace

# URL of the ESP32-CAM stream
url = 'http://192.168.4.1:81/stream'

# Create a VideoCapture object with the URL
cap = cv2.VideoCapture(url)

# Check if the camera is opened successfully
if not cap.isOpened():
    print("Error opening video stream or file")

# Load the reference image for face verification
img = cv2.imread("")  # Replace with your image path

# Variables to manage face matching and frame processing
frame_counter = 0
face_matched = False

def verify_face(frame):
    global face_matched
    try:
        # Verify the face using DeepFace
        result = DeepFace.verify(frame, img.copy())
        face_matched = result['verified']
    except ValueError:
        face_matched = False

while cap.isOpened():
    # Read a frame from the video capture
    ret, frame = cap.read()
    
    if ret:
        # Perform face verification every 30 frames
        if frame_counter % 30 == 0:
            threading.Thread(target=verify_face, args=(frame.copy(),)).start()
        
        frame_counter += 1

        # Display the verification result on the frame
        if face_matched:
            cv2.putText(frame, "MATCH!", (20, 450), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 255, 0), 3)
        else:
            cv2.putText(frame, "NO MATCH!", (20, 450), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 0, 255), 3)

        # Show the frame with the result
        cv2.imshow('ESP32-CAM Stream', frame)

    # Exit on 'q' key press
    if cv2.waitKey(1) == ord('q'):
        break

# Release resources and close windows
cap.release()
cv2.destroyAllWindows()
```
